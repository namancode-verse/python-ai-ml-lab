```python

import pandas as pd
df=pd.read_csv("German_Credit_Card.csv")
df
df.info()
print(df.shape)
df
```

```python
df.iloc[3:4,0:5]
```

```python
df['checkin_acc'].unique()  # it will give the unique catoegries in which we can classify the data,unique data's
```

```python
x_features=list(df.columns) # convertinginto list
x_features.remove('status')
encoded_df=pd.get_dummies(df[x_features],drop_first=True)
print(list(encoded_df.columns))
```

```python
encoded_df
```

```python
x=encoded_df
y=df['status']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier   #dept3 is optimal bec every col has 3 to 4 cateogries so the levels of the tree will b 3
clf=DecisionTreeClassifier(criterion="gini",max_depth=3)
clf.fit(x_train,y_train)        #here we will fit the training  x and y data to the decision tree
```

```python
pred_y=clf.predict(x_test)
```

```python
from sklearn import metrics
print("confusion matrix is \n",metrics.confusion_matrix(pred_y,y_test))
print("accuracy is",metrics.accuracy_score(pred_y,y_test))
print("auc score is",metrics.roc_auc_score(pred_y,y_test))
```
