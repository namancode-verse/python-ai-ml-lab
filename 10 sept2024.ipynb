```python
import numpy as np    
#to create an array 

# Create a sample 2D array
x = np.array(x)
y= np.array(y)

# Reshape x to be a column vector
x_reshaped = x.reshape(-1, 1)

print("Original shape:", x.shape)
print("Reshaped shape:", x_reshaped.shape)
```

```python
#split x and y into training and test data sets
#random_state--the set of data does not change
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
# print the dimensions of x_train,x_test,y_train,y_test
# print(x_train)

print(x_train.shape) # 70% in training,30% in testing and then random size =32 to keep the sample size same

print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lm=LinearRegression()

#tRAIN THE DATA model using training data sets
from sklearn.linear_model import LinearRegression
# Ensure x_train is 2D
# If x_train is already 2D, this step can be skipped
if x_train.ndim == 1:
    x_train = x_train.reshape(-1, 1)
if x_test.ndim == 1:
    x_test = x_test.reshape(-1, 1)

# Instantiate the model
lm = LinearRegression()


# Fit the model on the training data
lm.fit(x_train, y_train)


# Predict on the test set
y_pred = lm.predict(x_test)
```

```python
#visualising the test set results
plt.scatter(x_test,y_test,color="red")
plt.plot(x_test,)
```
